# Choosing reward values
We tried using different reward for the snake, first we started off with +1 for getting an apple -1 for dying.
Then we tried protecting the snake from dying by increasing the negative reward for dying to -10. 
But the snake wasnâ€™t getting good results so we tried adding an negative reward -0.5 for making a step with no apple on it.
But that didnâ€™t work as expected, so we went back to +1 for getting an apple -1 for dying.

| Option | Dead | Apple | Nothing | Amount of apples |
| --- | --- | --- | --- | --- |
| 1 | -1 | 1 | 0 | 1 |
| 2 | -10 | 1 | 0 | 1 |
| 3 | -10 | 1 | 1/grid size | 1 |
| **4** | **-1**  | **3** | **0** | 1 |
| 5 | -10 | 1 | 0.1 | 1 |
| 6 | -10 | 3 | 1/grid size | 1 |
| 7 | -1 | 3 | 1/grid size | 1 |
| --- | --- | --- | --- | --- |
| 8 | -1 | 3 | 0 | 3 |
| 9 | -1 | 3 | 1/grid * steps | 1 |

results:  
1: Noticeable low, wont get much higher than 0  
2: Good result, gets 4-5 apples and dies  
3: Comparable with 2  
4: Good result, gets 38-41 reward (13-14 apples, minus 1 for dead)  
5: Mediocre, between 1-2  
6: Gets a positive value, But doesn't get much above 2  
7: Decent result, has a minor tendency to run into walls  
8: Good result, similar to 4 but less targeted at apples  
9: Good result, similar to 4 and targets apples properly, though dies sooner
